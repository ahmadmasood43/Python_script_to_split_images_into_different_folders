```python
import cv2
import pandas as pd
import numpy as np
import os


input_dir = input("Please enter the full input data directory: ")
output_dir = input("Please enter the full output directory: ")
scv_dir = input("Enter the full directory for the CSV file :")

scv_path = scv_dir + '/MIDRC-RICORD-meta.csv'

typ_file = open("Typical.txt", "w+")
atyp_file = open("Atypical.txt", "w+")
ind_file = open("Indeterminate.txt", "w+")
neg_file = open("Negative.txt", "w+")

empty_tup = ("empt", 0)

column_read = ['fname','labels'] #list column headers that needs to be read

name_read = pd.read_csv(scv_path, usecols=column_read) #reads specified columns

name_list = name_read['fname'].to_numpy() #saving names of files in a separate array
label_list = name_read['labels'].to_numpy() #saving labels in separate arrays

label_range = len(label_list)
filename_range = len(name_list)

print(label_range + filename_range)

types_list = ["Negative", "Typical", "Indeterminate", "Atypical"]


for dir_count in types_list:
    parent_path = output_dir
    save_path = os.path.join(parent_path, dir_count)
    os.mkdir(save_path)

def high_cal(test):

    from collections import Counter
    freq = Counter(test.split()).most_common()   #calculates frequency of labels and saves them in a list
    range_counter = len(freq) #calculates size of list created
    cnt=0

    for cnt in range(range_counter): #we are supposed to classify them into 4 folders so the other labels are removed
        if freq[cnt][0] == 'Severe':
            freq[cnt] = empty_tup
        elif freq[cnt][0] == 'Mild' :
            freq[cnt] = empty_tup
        elif freq[cnt][0] == 'Moderate':
            freq[cnt] = empty_tup
        elif cnt == range_counter-1:
            break

    range_counter = len(freq) # the counter is set to the new length as many elements might get removed
    count = 0

    largest = freq[count][1]
    high_freq = freq[count][0]
    for count in range(range_counter): #loop runs so all elements are covered
        if freq[count][1] > largest:
            largest = freq[count][1]
            high_freq = freq[count][0]

    return high_freq



for count_name in range (label_range):


    if high_cal(label_list[count_name]) == "Typical":


        directory = "Typical"
        imagefile_dir = str(input_dir + "/" + name_list[count_name])
        image = cv2.imread(imagefile_dir)

        save_path = os.path.join(parent_path, directory)
        cv2.imwrite(os.path.join(save_path, name_list[count_name]), image)
        image_no = int(count_name + 2)

        typ_file.write(str(image_no) + " . " + (name_list[count_name] + '\n'))  # write to file

        print(str(image_no) + " . " + str(name_list[count_name]) + " == " + str(high_cal(label_list[count_name])))

    elif high_cal(label_list[count_name]) == "Atypical":


        directory = "Atypical"
        imagefile_dir = str(input_dir + "/" + name_list[count_name])

        image = cv2.imread(imagefile_dir)

        save_path = os.path.join(parent_path, directory)
        cv2.imwrite(os.path.join(save_path, name_list[count_name]), image)
        image_no = int(count_name + 2)

        atyp_file.write(str(image_no) + " . " + (name_list[count_name] + '\n'))  # write to file

        print(str(image_no) + " . " + str(name_list[count_name]) + " == " + str(high_cal(label_list[count_name])))

    elif high_cal(label_list[count_name]) == "Negative":

        directory = "Negative"
        imagefile_dir = str(input_dir + "/" + name_list[count_name])

        image = cv2.imread(imagefile_dir)

        save_path = os.path.join(parent_path, directory)
        cv2.imwrite(os.path.join(save_path, name_list[count_name]), image)
        image_no = int(count_name + 2)

        neg_file.write(str(image_no) + " . " + (name_list[count_name] + '\n'))  # write to file

        print(str(image_no) + " . " + str(name_list[count_name]) + " == " + str(high_cal(label_list[count_name])))

    elif high_cal(label_list[count_name]) == "Indeterminate":

        directory = 'Indeterminate'
        imagefile_dir = str(input_dir + "/" + name_list[count_name])

        image = cv2.imread(imagefile_dir)

        save_path = os.path.join(parent_path, directory)
        cv2.imwrite(os.path.join(save_path, name_list[count_name]), image)
        image_no = int(count_name + 2)

        ind_file.write(str(image_no) + " . " + (name_list[count_name] + '\n'))  # write to file

        print(str(image_no) + " . " + str(name_list[count_name]) + " == " + str(high_cal(label_list[count_name])))



ind_file.close()
neg_file.close()
typ_file.close()
atyp_file.close()
```
